# Volume Computation (Cut/Fill Analysis)

Compute earthwork volumes between two surfaces or against a reference elevation.

**Modules used:** `occulus.analysis.volume`

```python
import numpy as np
from occulus.types import PointCloud
from occulus.analysis import compute_volume
```

## Create two surfaces

Simulate a pre-mining and post-mining terrain surface.

```python
rng = np.random.default_rng(42)
n = 10_000
x = rng.uniform(0, 200, n)
y = rng.uniform(0, 200, n)

# Pre-mining: gentle hill
z_pre = 100 + 10 * np.sin(x / 50) * np.cos(y / 50) + rng.normal(0, 0.2, n)
surface_pre = PointCloud(np.column_stack([x, y, z_pre]))

# Post-mining: hill removed in central area (pit)
z_post = z_pre.copy()
pit_mask = ((x - 100)**2 + (y - 100)**2) < 50**2
z_post[pit_mask] -= 5.0  # 5m excavation
surface_post = PointCloud(np.column_stack([x, y, z_post]))

print(f"Pre-mining:  {surface_pre.n_points:,} points, Z range: {z_pre.min():.1f}–{z_pre.max():.1f} m")
print(f"Post-mining: {surface_post.n_points:,} points, Z range: {z_post.min():.1f}–{z_post.max():.1f} m")
```

## Compute volume difference

```python
result = compute_volume(surface_pre, surface_post, resolution=2.0)

print(f"Cut volume:  {result.cut_volume:,.0f} m³")
print(f"Fill volume: {result.fill_volume:,.0f} m³")
print(f"Net volume:  {result.net_volume:,.0f} m³")
print(f"Total area:  {result.area:,.0f} m²")
print(f"Cut area:    {result.cut_area:,.0f} m²")
print(f"Fill area:   {result.fill_area:,.0f} m²")
```

## Volume against a flat reference

```python
# How much material is above 100m elevation?
result_flat = compute_volume(surface_pre, 100.0, resolution=2.0)
print(f"Volume above 100m: {result_flat.cut_volume:,.0f} m³")
print(f"Volume below 100m: {result_flat.fill_volume:,.0f} m³")
```
