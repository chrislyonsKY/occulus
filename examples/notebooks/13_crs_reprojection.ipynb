# CRS Reprojection with Occulus

Reproject point clouds between coordinate reference systems using `occulus.crs`.

**Modules used:** `occulus.crs`, `occulus.io`

```python
import numpy as np
from occulus.types import PointCloud, AcquisitionMetadata
from occulus.crs import reproject, transform_coordinates
```

## Create a sample cloud in UTM Zone 17N

```python
# Simulated survey points near Frankfort, KY (UTM 17N)
rng = np.random.default_rng(42)
xyz = np.column_stack([
    rng.uniform(625_000, 626_000, 500),  # easting
    rng.uniform(4_225_000, 4_226_000, 500),  # northing
    rng.uniform(180, 220, 500),  # elevation
])

meta = AcquisitionMetadata(coordinate_system="EPSG:32617")
cloud = PointCloud(xyz, metadata=meta)
print(f"Points: {cloud.n_points}")
print(f"Source CRS: {cloud.metadata.coordinate_system}")
print(f"Bounds: {cloud.bounds}")
```

## Reproject to WGS84 (EPSG:4326)

```python
wgs84 = cloud.reproject("EPSG:4326")
print(f"Target CRS: {wgs84.metadata.coordinate_system}")
print(f"Bounds (lon/lat): {wgs84.bounds}")
print(f"\nFirst 3 points (UTM):")
print(cloud.xyz[:3])
print(f"\nFirst 3 points (WGS84):")
print(wgs84.xyz[:3])
```

## Low-level coordinate transform

```python
# Transform just the coordinates without creating a new cloud
lonlat = transform_coordinates(xyz[:5], "EPSG:32617", "EPSG:4326")
print("UTM -> WGS84:")
for i in range(5):
    print(f"  ({xyz[i,0]:.0f}, {xyz[i,1]:.0f}) -> ({lonlat[i,0]:.6f}, {lonlat[i,1]:.6f})")
```
